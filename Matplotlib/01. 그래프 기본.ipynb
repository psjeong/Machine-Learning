# Matplotlib

```python
import matplotlib.pyplot as plt
```

# 1.그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
```

```python
print(plt.plot(x, y))
```

```python
plt.plot(x, y)
plt.show()  
```

## Title 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
# -> 한글깨짐
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 포트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist]
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
```

```python
plt.plot([-1, 0, 1],[-5, -1, 2])
# '-(마이너스)'부분이 폰트 깨짐
```

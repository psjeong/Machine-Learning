# 2.축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 포트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family': 'HYGungso-Bold', 'size':20}) # 개별 폰트 설정
```

```python
plt.plot(x,y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
# 컬러 적용
plt.plot(x,y)
plt.xlabel('X축', color='red')
plt.ylabel('Y축', color="#00aa00") # rgb 가능
```

```python
# 위치 변경
plt.plot(x,y)
plt.xlabel('X축', color='red', loc='right')  # left, center, right
plt.ylabel('Y축', color="#00aa00",loc='top') # top, center, bottom
```

```python
# ticks 간소화 가능
plt.plot(x,y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```

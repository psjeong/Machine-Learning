# 03. 범례(legend)


```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 포트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, label='데이터')
plt.legend()
```

**범례 위치 변경** <br>
- best : defualt 값 | 데이터를 확인후 데이터를 가리지 않게 위치 시킴
- upper right
- upper left
- lower right
- lower left
- right
- center left
- center right
- lower center
- upper center
- center

-참고 사이트 : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```python
plt.plot(x, y, label='데이터')
plt.legend(loc='upper right') # 범례 위치 변경(위)
```

```python
plt.plot(x, y, label='데이터')
plt.legend(loc='lower right') # 범례 위치 변경(아래)
```

```python
# x, y축으로 원하는 위치 지정
plt.plot(x, y, label='범례')
plt.legend(loc=(0.5,0.5)) # loc=(x축,y축) -> 0~1 사이 값
```

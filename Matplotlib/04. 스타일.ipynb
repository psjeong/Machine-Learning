# 04. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 포트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

**기본 스타일**

```python
plt.plot(x, y)
```

**라인 두께 조절**

```python
plt.plot(x, y, linewidth=5)
```

## 마커(marker)
maker 모양 참고 : https://matplotlib.org/stable/api/markers_api.html
- marker='o'

```python
plt.plot(x, y, marker='o')
```

- 라인 지우기

```python
plt.plot(x, y, marker='o', linestyle='None')
```

- marker = 'v'

```python
plt.plot(x, y, marker='v')
```

- marker 사이즈 조정

```python
plt.plot(x, y, marker='v', markersize=10)
```

- marker = 'X'

```python
plt.plot(x, y, marker='X', markersize=10)
```

- marker 색상 조정

```python
# marker의 겉둘레 컬러 변경(markeredgecolor)
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red')
```

```python
# marker 컬러 변경(markerfacecolor)
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일

참고 : https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

```python
plt.plot(x, y, linestyle='-') #  linestyle='-' : default 값으로 생략 가능
```

## 색깔
참고 : https://matplotlib.org/stable/gallery/color/named_colors.html

```python
plt.plot(x, y, color='pink')
```

```python
# rgb 값 적용
plt.plot(x, y, color='#ff0000')
```

```python
# 'b' : blue (상단에 있는 사이트 참고)
plt.plot(x, y, color='b')
```

```python
plt.plot(x, y, color='g')
```

## 포맷

```python
plt.plot(x, y, 'ro--') # color(r), marker(o), linestyle(--)
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go') 
# plt.plot(x, y, color='g', marker='o', linestyle='None') 같은 결과
```

## 축약어
참고 : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
# mfc (markerfacecolor)
# ms (markersize)
# mec (markeredgecolor)
# ls (linestyle)
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.3) # alpha :  투명도 (0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
# dpi : 인치당 도트 수. 즉, 해상도(확대)
plt.figure(figsize=(10, 5), dpi = 150)
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x, y)
```
